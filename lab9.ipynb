```python
# Imports

import pygame, sys
from pygame.locals import *
import random, time
 
# Initialzing 

pygame.init()
 
# Setting up FPS
 
FPS = 60
FramePerSec = pygame.time.Clock()
 
# Creating colors

BLUE  = (0, 0, 255)
RED   = (255, 0, 0)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
 
# Other Variables for use in the program

SCREEN_WIDTH = 400
SCREEN_HEIGHT = 600
SPEED = 5
SCORE = 0
COINS = 0

COIN_THRESHOLD = 5       # Increase enemy speed every 5 coins
last_coin_level = 0      # Keeps track of how many thresholds passed


SPEED1 = 5
 
# Setting up Fonts

font = pygame.font.SysFont("Verdana", 60)
# font_medium = pygame.font.SysFont("Verdana", 40)
font_small = pygame.font.SysFont("Verdana", 20)
game_over = font.render("Game Over", True, BLACK)
# score_text1 = font_medium.render(f"Your Score: {SCORE}", True, BLACK)
# coins_text1 = font_medium.render(f"Total Coins: {COINS}", True, BLACK)
 
background = pygame.image.load("AnimatedStreet.png")
 
# Create a white screen
 
DISPLAYSURF = pygame.display.set_mode((400,600))
DISPLAYSURF.fill(WHITE)
pygame.display.set_caption("Game")
 
class Enemy(pygame.sprite.Sprite):
      def __init__(self):
        super().__init__() 
        self.image = pygame.image.load("Enemy.png")
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)  
 
      def move(self):
        global SCORE
        self.rect.move_ip(0, SPEED)
        if (self.rect.top > 600):
            SCORE += 10
            self.rect.top = 0
            self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)
 
 
class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__() 
        self.image = pygame.image.load("Player.png")
        self.rect = self.image.get_rect()
        self.rect.center = (160, 520)
        
    def move(self):
        pressed_keys = pygame.key.get_pressed()
       # if pressed_keys[K_UP]:
            # self.rect.move_ip(0, -5)
       # if pressed_keys[K_DOWN]:
            #s elf.rect.move_ip(0,5)
         
        if self.rect.left > 0:
              if pressed_keys[K_LEFT]:
                  self.rect.move_ip(-5, 0)
        if self.rect.right < SCREEN_WIDTH:        
              if pressed_keys[K_RIGHT]:
                  self.rect.move_ip(5, 0)

class Coin(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.weights = {
            "CoinBronze.png": 1,
            "CoinSilver.png": 3,
            "CoinGold.png": 5
        }

        # choose random coin from the list

        self.image_name = random.choice(list(self.weights.keys()))
        self.image = pygame.image.load(self.image_name)
        self.image = pygame.transform.scale_by(self.image, 0.1)
        self.weight = self.weights[self.image_name]
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)

    def move(self):
        self.rect.move_ip(0, SPEED1)
        if self.rect.top > 600:
            self.reset_position()

    # reset position after collision with coin

    def reset_position(self):
        self.image_name = random.choice(list(self.weights.keys()))
        self.image = pygame.image.load(self.image_name)
        self.image = pygame.transform.scale_by(self.image, 0.1)
        self.weight = self.weights[self.image_name]
        self.rect = self.image.get_rect(center=(random.randint(40, SCREEN_WIDTH - 40), 0))

                   
# Setting up Sprites
        
P1 = Player()
E1 = Enemy()
C1 = Coin()
 
# Creating Sprites Groups

enemies = pygame.sprite.Group()
enemies.add(E1)
coins1 = pygame.sprite.Group()
coins1.add(C1)
all_sprites = pygame.sprite.Group()
all_sprites.add(P1)
all_sprites.add(E1)
all_sprites.add(C1)
 
# Adding a new User event 

# INC_SPEED = pygame.USEREVENT + 1
# pygame.time.set_timer(INC_SPEED, 1000)
 
# Game Loop

while True:
       
    # Cycles through all events occurring
      
    for event in pygame.event.get():
        # if event.type == INC_SPEED:
            # SPEED += 0.5     
        if event.type == QUIT:
            pygame.quit()
            sys.exit()
    
    # Display elements of the screen
    
    DISPLAYSURF.blit(background, (0, 0))
    score_text = font_small.render("Score: " + str(SCORE), True, BLACK)
    coins_text = font_small.render("Coins: " + str(COINS), True, BLACK)
    DISPLAYSURF.blit(score_text, (10, 10))
    DISPLAYSURF.blit(coins_text, (10, 30))

    # Moves and Re-draws all Sprites
    
    for entity in all_sprites:
        DISPLAYSURF.blit(entity.image, entity.rect)
        entity.move()

    # To be run if collision occurs between Player and Coin
    
    if pygame.sprite.spritecollideany(P1, coins1):
        pygame.mixer.Sound("coin_sound.wav").play()
        for coin in coins1:
            if pygame.sprite.collide_rect(P1, coin):  # ensure collision is with specific coin
                COINS += coin.weight                  # use coin's weight
                coin.reset_position()                 # use method to reset coin

                # Check if we crossed a threshold

                if COINS // COIN_THRESHOLD > last_coin_level:
                    last_coin_level = COINS // COIN_THRESHOLD
                    SPEED += 1  # Increase enemy speed
        
    # To be run if collision occurs between Player and Enemy

    if pygame.sprite.spritecollideany(P1, enemies):
        pygame.mixer.Sound('crash.wav').play()
        time.sleep(0.5)
                    
        DISPLAYSURF.fill(RED)
        DISPLAYSURF.blit(game_over, (30, 250))
        # DISPLAYSURF.blit(score_text1, (50, 330))
        # DISPLAYSURF.blit(coins_text1, (50, 380))
           
        pygame.display.update()
        for entity in all_sprites:
            entity.kill() 
        time.sleep(2)
        pygame.quit()
        sys.exit()        
         
    pygame.display.update()
    FramePerSec.tick(FPS)
```

Snake

```python
import pygame # type: ignore
import random

pygame.init()

width = 800
height = 600
screen = pygame.display.set_mode((width, height))

# Variables and their initialization

level = 0
speed = 200
fruits_eaten = 0
score = 0
fruit_eaten = False

# Gold apple

gold_active = False
gold_timer = 0
gold_coor = [0, 0]
gold_weight = 50

# Spawn the first fruit

fr_x = random.randrange(1, width // 10) * 10
fr_y = random.randrange(1, height // 10) * 10
fruit_coor = [fr_x, fr_y]

# Snake's body

head_square = [100, 100]

squares = [
    [80,100],
    [90,100],
    [100,100]
]

direction = "right"
next_dir = "right"

# Game over message

done = False
def game_over(font, size, color):
    global done
    g_o_font = pygame.font.SysFont(font, size)
    g_o_surface = g_o_font.render("Game Over, your score: " + str(score), True, color)
    g_o_rect = g_o_surface.get_rect(center = (width // 2, height // 2))

    screen.blit(g_o_surface, g_o_rect)
    pygame.display.update()

    pygame.time.delay(4000)
    pygame.quit()

# Start of gameplay loop

while not done:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_DOWN:
                next_dir = "down"
            if event.key == pygame.K_UP:
                next_dir = "up"
            if event.key == pygame.K_LEFT:
                next_dir = "left"
            if event.key == pygame.K_RIGHT:
                next_dir = "right"

    # Game over conditions

    for square in squares[:-1]:
        if head_square[0] == square[0] and head_square[1] == square[1]:
            game_over("times new roman", 45, (128, 128, 128))

    if head_square[0] < 0 or head_square[0] >= width or head_square[1] < 0 or head_square[1] >= height:
        game_over("times new roman", 45, (128, 128, 128))
    
    # Change of Direction

    if next_dir == "right" and direction != "left":
        direction = "right"
    if next_dir == "up" and direction != "down":
        direction = "up"
    if next_dir == "left" and direction != "right":
        direction = "left"
    if next_dir == "down" and direction != "up":
        direction = "down"

    if direction == "right":
        head_square[0] += 10
    if direction == "left":
        head_square[0] -= 10
    if direction == "up":
        head_square[1] -= 10
    if direction == "down":
        head_square[1] += 10

    new_square = [head_square[0], head_square[1]]

    squares.append(new_square)
    squares.pop(0)

    # When fruit is eaten

    if head_square[0] == fruit_coor[0] and head_square[1] == fruit_coor[1]:
        fruit_eaten = True
        score +=10

        fruits_eaten += 1

        # Grow body

        squares.insert(0, squares[0][:])

        # Change of Speed

        if fruits_eaten % 3 == 0:
            level += 1
            speed = max(50, speed - 20)
    
    # Spawn new fruit

    if fruit_eaten:

        fr_x = random.randrange(1, width // 10) * 10
        fr_y = random.randrange(1, height // 10) * 10
        fruit_coor = [fr_x, fr_y]
        fruit_eaten = False

    # Gold apple eaten

    if gold_active and head_square[0] == gold_coor[0] and head_square[1] == gold_coor[1]:
        score += gold_weight
        fruits_eaten += 1
        gold_active = False
        squares.insert(0, squares[0][:])  # Grow body

        if fruits_eaten % 3 == 0:
            level += 1
            speed = max(50, speed - 20)

    # Randomly spawn gold apple (10% chance)

    if not gold_active and random.randint(1, 100) <= 10:
        gold_x = random.randrange(1, width // 10) * 10
        gold_y = random.randrange(1, height // 10) * 10
        gold_coor = [gold_x, gold_y]
        gold_timer = pygame.time.get_ticks()
        gold_active = True

    # Remove gold apple after 10 seconds

    if gold_active and pygame.time.get_ticks() - gold_timer > 10000:
        gold_active = False

    # Drawing section

    screen.fill((0, 0, 0))

    score_font = pygame.font.SysFont("times new roman", 20)
    score_surface = score_font.render(f"Score: {score} Level: {level}", True, (128, 128, 128))
    score_rect = score_surface.get_rect()

    screen.blit(score_surface,score_rect)

    # Regular apple

    if not fruit_eaten:
        pygame.draw.circle(screen, (255, 0, 0), (fruit_coor[0] + 5, fruit_coor[1] + 5), 5)

    # Gold apple

    if gold_active:
        pygame.draw.circle(screen, (255, 215, 0), (gold_coor[0] + 5, gold_coor[1] + 5), 6)

        # Draw timer above the gold apple

        time_left = max(0, 10 - (pygame.time.get_ticks() - gold_timer) // 1000)
        gold_font = pygame.font.SysFont("times new roman", 16)
        gold_timer_surface = gold_font.render(str(time_left), True, (255, 215, 0))
        gold_timer_rect = gold_timer_surface.get_rect(center=(gold_coor[0] + 5, gold_coor[1] - 10))
        screen.blit(gold_timer_surface, gold_timer_rect)



    for el in squares:
        pygame.draw.rect(screen, (0, 255, 64),
                         pygame.Rect(el[0], el[1], 10, 10))

    pygame.display.flip()
    pygame.time.delay(speed)

pygame.quit()
```

```python
import math
import pygame

def main():
    pygame.init()
    screen = pygame.display.set_mode((640, 480))
    clock = pygame.time.Clock()
    
    # Variables

    radius = 15
    mode = 'blue'
    drawing = False
    erasing = False
    shape_mode = None
    last_pos = None

    canvas = pygame.Surface(screen.get_size())
    canvas = canvas.convert()
    canvas.fill((0, 0, 0))

    running = True

    # Main loop

    while running:
        pressed = pygame.key.get_pressed()

        alt_held = pressed[pygame.K_LALT] or pressed[pygame.K_RALT]
        ctrl_held = pressed[pygame.K_LCTRL] or pressed[pygame.K_RCTRL]

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

            # Controls:
            #  R/G/B - Color Red / Green / Blue
            #  Z - Rectangle
            #  X - Circle
            #  S - Square
            #  T - Right Triangle
            #  E - Equilateral Triangle
            #  H - Rhombus
            #  LMB - Drawing shape / free draw
            #  RMB - Erasing
            #  -/+ - Changing brush radius
            
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_w and ctrl_held:
                    return
                if event.key == pygame.K_F4 and alt_held:
                    return
                if event.key == pygame.K_ESCAPE:
                    return
                
                elif event.key == pygame.K_s:
                    shape_mode = 'square'            # Square
                elif event.key == pygame.K_t:
                    shape_mode = 'right_triangle'    # Right triangle
                elif event.key == pygame.K_e:
                    shape_mode = 'equilateral'       # Equilateral triangle
                elif event.key == pygame.K_h:
                    shape_mode = 'rhombus'           # Rhombus

                
                # Change color mode

                if event.key == pygame.K_r:
                    mode = 'red'
                elif event.key == pygame.K_g:
                    mode = 'green'
                elif event.key == pygame.K_b:
                    mode = 'blue'
                elif event.key == pygame.K_x:
                    shape_mode = 'circle'
                elif event.key == pygame.K_z:
                    shape_mode = 'rect'

                # Change brush radius

                elif event.key == pygame.K_EQUALS or event.key == pygame.K_PLUS:
                    radius = min(200, radius + 1)
                elif event.key == pygame.K_MINUS:
                    radius = max(1, radius - 1)

            if event.type == pygame.MOUSEBUTTONDOWN:
                if event.button == 1:
                    drawing = True
                    erasing = False
                    last_pos = event.pos
                    
                    if shape_mode == 'circle':
                        pygame.draw.circle(canvas, getColor(mode), event.pos, radius)
                    elif shape_mode == 'rect':
                        rect = pygame.Rect(0, 0, radius * 2, radius * 2)
                        rect.center = event.pos
                        pygame.draw.rect(canvas, getColor(mode), rect)
                    elif shape_mode == 'square':
                        draw_square(canvas, getColor(mode), event.pos, radius)
                    elif shape_mode == 'right_triangle':
                        draw_right_triangle(canvas, getColor(mode), event.pos, radius)
                    elif shape_mode == 'equilateral':
                        draw_equilateral_triangle(canvas, getColor(mode), event.pos, radius)
                    elif shape_mode == 'rhombus':
                        draw_rhombus(canvas, getColor(mode), event.pos, radius)
                    shape_mode = None


            if event.type == pygame.MOUSEBUTTONUP:
                if event.button in (1, 3):
                    drawing = False
                    erasing = False
                    last_pos = None

            if event.type == pygame.MOUSEMOTION:
                if drawing and last_pos:
                    color = (0, 0, 0) if erasing else getColor(mode)

                    # Draw smooth continuous line

                    dx = event.pos[0] - last_pos[0]
                    dy = event.pos[1] - last_pos[1]
                    distance = max(abs(dx), abs(dy))

                    for i in range(distance):
                        x = int(last_pos[0] + float(i) / distance * dx)
                        y = int(last_pos[1] + float(i) / distance * dy)
                        pygame.draw.circle(canvas, color, (x, y), radius)

                    last_pos = event.pos
                elif drawing:
                    last_pos = event.pos

        screen.blit(canvas, (0, 0))
        pygame.display.flip()
        clock.tick(60)

    pygame.quit()

def getColor(color_mode):
    if color_mode == 'blue':
        return (0, 0, 255)
    elif color_mode == 'red':
        return (255, 0, 0)
    elif color_mode == 'green':
        return (0, 255, 0)
    return (255, 255, 255)

# Draws a square centered at `pos` with side length `2*radius`

def draw_square(surface, color, pos, radius):
    side = 2 * radius
    rect = pygame.Rect(0, 0, side, side)
    rect.center = pos
    pygame.draw.rect(surface, color, rect)

# Draws a right triangle centered at `pos`

def draw_right_triangle(surface, color, pos, radius):
    x, y = pos
    points = [
        (x, y - radius),       # top
        (x - radius, y + radius),  # bottom left
        (x + radius, y + radius)   # bottom right
    ]
    pygame.draw.polygon(surface, color, points)

# Draws an equilateral triangle with a flat base, centered at `pos`

def draw_equilateral_triangle(surface, color, pos, radius):
    x, y = pos
    side = 2 * radius
    height = math.sqrt(3) / 2 * side

    # Triangle points (flat base)
    points = [
        (x, y - height / 2),                 # Top
        (x - side / 2, y + height / 2),      # Bottom left
        (x + side / 2, y + height / 2)       # Bottom right
    ]
    pygame.draw.polygon(surface, color, points)

# Draws a rhombus centered at `pos` with diagonals `2*radius`

def draw_rhombus(surface, color, pos, radius):
    x, y = pos
    points = [
        (x, y - radius),  # top
        (x + radius, y),  # right
        (x, y + radius),  # bottom
        (x - radius, y)   # left
    ]
    pygame.draw.polygon(surface, color, points)

main()
```
